```python
import numpy as np
from sklearn.datasets import load_diabetes #scikit-learn is a machine learning library in Python that provides various tools for data analysis and modeling. The load_diabetes function is used to load the diabetes dataset, which is a commonly used dataset for regression tasks.
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

Here is a breakdown of the specific arguments passed into the function:

X, y: These are the feature matrix (X) and the target vector (y) that you previously loaded. They are passed in together so that their rows are shuffled and split in sync, ensuring that each feature row in X still matches up with its correct target label in y.

test_size=0.2: This determines the proportion of the dataset to include in the test split. Passing 0.2 allocates 20% of the data for testing, leaving the remaining 80% for training the model.

random_state=2: Before splitting, the function randomly shuffles the data. Providing a specific integer seed (like 2) controls this randomization process. This ensures reproducibility; every time you run this cell, the data will be split in exactly the same way. If random_state=None (the default), the split is different each run; any integer can be used as a seed (common choices are 0, 1, or 42).

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

What it is: r2_score computes the coefficient of determination for a regression model — a single-number summary of how well predictions match the true targets.

Formula: R² = 1 − (Σ(y_true − y_pred)²) / (Σ(y_true − y_mean)²). It measures the proportion of variance in y_true explained by the model.

Range & meaning: 1.0 = perfect fit; 0.0 = model is as good as predicting the mean; negative values = model is worse than predicting the mean. Higher is better.

When to use: Use for regression tasks to compare models or to report goodness-of-fit. Be cautious with outliers, non-linear relationships, or when target variance is tiny.

Usage example: r2_score(y_test, y_pred) returns the R² for those arrays.

```python
class MultipleLinearRegression:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self, X_train, y_train):
        X_train = np.insert(X_train, 0, 1, axis=1) #added a column of 1s to the left of X_train for the intercept term
        
        #calculate the coefficients
        betas = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]
        

    def predict(self, X_test):
        y_pred = np.dot(X_test, self.coef_) + self.intercept_
        return y_pred
```

```python
mlr = MultipleLinearRegression()
```

```python
mlr.fit(X_train, y_train)
```

```python
mlr_y_pred = mlr.predict(X_test)
print(mlr_y_pred)
```
